```python
import pandas as pd
```

```python
DATA_PATH = 'yt_clean.csv'
NONSENSE_WORD_PATH = 'nonsense_word.txt'
KEY_WORD_PATH = 'key_word.txt'
data = pd.read_csv(DATA_PATH)
data
```

```python
#delete nonsense data
with open(NONSENSE_WORD_PATH, 'r', encoding='utf-8') as file:
    nonsense_word = [line.strip() for line in file.readlines()]


pattern = '|'.join(nonsense_word)

# Drop rows that contain any of the nonsense words in a specific column
data = data[~data['data'].str.contains(pattern, na=False)]
data
```

```python
data_sorted = data.sort_values(by='comment',ascending=False)
data_sorted
```

```python
#find key data
with open(KEY_WORD_PATH, 'r', encoding='utf-8') as file:
    key_word = [line.strip() for line in file.readlines()]

pattern = '|'.join(key_word)

# Keep rows where 'data' contains any of the keywords
data = data[data['data'].str.contains(pattern, na=False)]
data = data.sort_values(by='likes',ascending=False)
data
```
